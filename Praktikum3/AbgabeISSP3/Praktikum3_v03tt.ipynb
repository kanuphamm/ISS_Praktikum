Erstellen Sie Python-Scripte für die Vorverarbeitung Ihrer Messdaten mittels Filter.

Verwenden Sie für die Anwendung eines Filters den Datensatz für das Bewegungsprofil 2 aus dem letzten Praktikum.

Filtern Sie die Daten von einem der Beschleunigungssenoren für alle drei Achsen mit jeweils einem IIR-Filter erster Ordnung (siehe Vorlesung); mindestens 1000 Samples. Implementieren Sie das Filter so, dass man es auch für einen Live-Betrieb (jeweils ein Sample pro Schleifendurchlauf) verwendet werden kann.

**Code**

```python
import numpy as np
import matplotlib.pyplot as plt

data_x = np.genfromtxt("DataSets/Datensatz2.csv", delimiter=";", usecols=(7,), skip_header=1)
data_y = np.genfromtxt("DataSets/Datensatz2.csv", delimiter=";", usecols=(8,), skip_header=1)
data_z = np.genfromtxt("DataSets/Datensatz2.csv", delimiter=";", usecols=(9,), skip_header=1)

# Größe des Eingangssignals ermitteln
#length = data_x.shape[0]
b0 = 0
a1 = 1

def iir_filter(data, b0, a1):
    length = data.shape[0]
    g = np.zeros(length)
    xk1 = 0

    for i in range(length):
        s = data[i]
        xk1 = -a1 * xk1 + s
        g[i] = b0 * xk1

    return g, length


fil_x, length = iir_filter(data_x, b0, a1)
fil_y, length = iir_filter(data_y, b0, a1)
fil_z, length = iir_filter(data_z, b0, a1)

# Anpassung der x-Achse
time_in_seconds = length / 162
x = np.arange(0, time_in_seconds, 1/162)
plt.figure(figsize=(10, 6))

plt.plot(x, data_y, color='#ffb0ab', label='Origin y')
plt.plot(x, data_x, color='#badaff', label='Origin x')
plt.plot(x, data_z, color='#c2ffc8', label='Origin z')
plt.plot(x, fil_y, color='#ff1100', label='Filtered y')


plt.plot(x, fil_x, color='#3895ff', label='Filtered x')


plt.plot(x, fil_z, color='#00ff1a', label='Filtered z')

plt.title("Datensatz 2: Beschleunigungsachsen des LIS2DW12")
plt.xlabel("t [s]")
plt.ylabel("[acc/mg]")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler


# Daten einlesen
data4 = np.genfromtxt("DataSets/Datensatz4.csv", delimiter=";", skip_header=1)
data4 = data4[:,1:]
# NaN-Werte entfernen
data4 = np.nan_to_num(data4)

scaler_MM = MinMaxScaler(feature_range=(-1, 1))
scaler_MM.fit(data4) 
normal_data4 = scaler_MM.transform(data4)

# PCA durchführen
pca4n = PCA()
pca4n.fit(normal_data4)
normal_data4 = pca4n.transform(normal_data4)

# Eigenwerte berechnen
eigenvalues = pca4n.explained_variance_
eigenvalues = np.array(eigenvalues)

# Scree-Plot erstellen
plt.plot(np.arange(1, len(eigenvalues) + 1), eigenvalues, marker='o')
plt.xlabel('Hauptkomponente')
plt.ylabel('Eigenwert')
plt.title('Scree-Plot Eigenwerte DS4')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Daten einlesen
data = np.genfromtxt("Datasets/Datensatz6.csv", delimiter=";", skip_header=1)
data = data[:,1:]
# NaN-Werte entfernen
data = np.nan_to_num(data)

# PCA durchführen
pca = PCA()
pca.fit(data)
data = pca.transform(data)

# Eigenwerte berechnen
eigenvalues = pca.explained_variance_
eigenvalues = np.array(eigenvalues)

# Scree-Plot erstellen
plt.plot(np.arange(1, len(eigenvalues) + 1), eigenvalues, marker='o')
plt.xlabel('Hauptkomponente')
plt.ylabel('Eigenwert')
plt.title('Scree-Plot Eigenwerte DS6')
plt.show()
```
